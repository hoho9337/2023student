```python
import pandas as pd
import numpy as np
```

```python
with pd.ExcelFile("2023student.xlsx") as xls:
    df = pd.read_excel(xls, "Sheet1")

df_stu_name1 = df.iloc[:,[8,5,6,10]]
df_stu_name1 = df_stu_name1.dropna(subset=[df.columns[5]])
df_stu_name1.columns = ['학년','이름','세례명','학생번호']
df_stu_name2 = df.iloc[:,[17,11,13,21]]
df_stu_name2 = df_stu_name2.dropna(subset=[df.columns[11]])
df_stu_name2.columns = ['학년','이름','세례명','학생번호']
df_stu_name3 = df.iloc[:,[18,12,14,22]]
df_stu_name3 = df_stu_name3.dropna(subset=[df.columns[12]])
df_stu_name3.columns = ['학년','이름','세례명','학생번호']
df_stu_name4 = df.iloc[:,[32,23,26,38]]
df_stu_name4 = df_stu_name4.dropna(subset=[df.columns[23]])
df_stu_name4.columns = ['학년','이름','세례명','학생번호']
df_stu_name5 = df.iloc[:,[33,24,27,39]]
df_stu_name5 = df_stu_name5.dropna(subset=[df.columns[24]])
df_stu_name5.columns = ['학년','이름','세례명','학생번호']
df_stu_name6 = df.iloc[:,[34,25,28,40]]
df_stu_name6 = df_stu_name6.dropna(subset=[df.columns[25]])
df_stu_name6.columns = ['학년','이름','세례명','학생번호']

names = [df_stu_name1,df_stu_name2,df_stu_name3,df_stu_name4,df_stu_name5,df_stu_name6]
df_stu_name = pd.concat(names,ignore_index=False)
df_stu_name = df_stu_name.fillna(0)

df_stu_name = df_stu_name.astype({'학생번호':'str'})
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace("-","")
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace(" ","")
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace("\.0","")
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace("없음","0")
df_stu_name = df_stu_name.astype({'학생번호':'int'})
df_stu_name = df_stu_name.astype({'학생번호':'str'})
df_stu_name
```

```python
df_stu_num = df_stu_name.copy()
d = {'중1': 1,
     '중2': 2,
     '중3': 3,
     '고1': 4,
     '고2': 5,
     '고3': 6,
     '수험생':7,
    }

df_stu_num['학년'] = df_stu_num['학년'].map(d)
df_stu_num = df_stu_num.sort_values(by='학년',na_position='first')

b = {v:k for k,v in d.items()}

df_stu_num['학년'] = df_stu_num['학년'].map(b)
df_stu_num
```

```python
df_all = df_stu_num.copy()
idx = df_all.index
df_all = pd.concat([df_all,df.iloc[:,[3,4]]],axis=1)
df_all = df_all.rename({df_all.columns[5]:'학부모님 번호'},axis='columns')

df_stu_name = df_stu_name.astype({'학생번호':'str'})
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace("-","")
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace(" ","")
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace("\.0","")
df_stu_name["학생번호"] = df_stu_name['학생번호'].str.replace("없음","0")
df_stu_name = df_stu_name.astype({'학생번호':'int'})
df_stu_name

df_all = df_all.astype({df_all.columns[5]:'str'})
df_all["학부모님 번호"] = df_all['학부모님 번호'].str.replace("-","")
df_all["학부모님 번호"] = df_all['학부모님 번호'].str.replace(" ","")
df_all = df_all.astype({df_all.columns[5]:'int'})
df_all = df_all.astype({df_all.columns[5]:'str'})

df_all = df_all.reset_index(drop=True)

df_all[df_all.columns[3]] = df_all[df_all.columns[3]].apply(lambda x : '0'+ x)
df_all[df_all.columns[5]] = df_all[df_all.columns[5]].apply(lambda x : '0'+ x)
df_all
```

```python
df_all.to_excel('학생명단.xlsx')
```
